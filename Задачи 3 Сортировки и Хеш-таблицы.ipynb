```python
#Найти корень числа(ближайшее целое)

def binary_search_sqrt(target: int) -> int:
    l, r = 0, target
    while l <= r:
        middle = (l + r) // 2  
        square = middle * middle
        if square == target:
            return middle
        elif square < target:
            l = middle + 1
        else:
            r = middle - 1
    return r  
```

```python
#Очень легкая задача(про 2 ксерокса)

def min_copy_time(n: int, x: int, y: int) -> int:
    if n == 1:
        return min(x, y) 
    l, r = 0, (n - 1) * max(x, y)
    while l + 1 < r:
        mid = (l + r) // 2
        copies = mid // x + mid // y
        if copies < n - 1:
            l = mid
        else:
            r = mid
    return r + min(x, y)

print(min_copy_time(5, 1, 2))  
print(min_copy_time(4, 2, 3))  
print(min_copy_time(10, 1, 1))
```

```python
#Накормить животных
#похожая ксати лежит на литкоде №455

def feed_animals(animals: list[int], food: list[int]) -> int:
    if not animals or not food:
        return 0
    animals.sort()
    food.sort()
    count = 0
    for portion in food:
        if count < len(animals) and portion >= animals[count]:
            count += 1
        if count == len(animals):
            break
    return count

print(feed_animals([3, 4, 7], [8, 1, 2]))      # 1
print(feed_animals([3, 8, 1, 4], [1, 1, 2]))   # 1
print(feed_animals([1, 2, 2], [7, 1]))         # 2
print(feed_animals([8, 2, 3, 2], [1, 4, 3, 8]))# 3
```

```python
#Найти разницу между двух строк

def extra_letter(a: str, b: str) -> str:
    hash_b = {}
    for ch in b:
        hash_b[ch] = hash_b.get(ch, 0) + 1
    for ch in a:
        if ch in hash_b:
            hash_b[ch] -= 1
    for ch, count in hash_b.items():
        if count > 0:
            return ch
    return "Разницы нет"  



print(extra_letter("uio", "oeiu"))  # e
print(extra_letter("fe", "efo"))    # o
print(extra_letter("ab", "ab"))     # ""
print(extra_letter("bbb", "bbbb"))  # b
```

```python
#альтернатива решения через коды символов
#считаем сумму кодов в первой и второй строке и раздница-код добавленного символа
def extra_letter_fast(a: str, b: str) -> str:
    return chr(sum(map(ord, b)) - sum(map(ord, a)))

print(extra_letter_fast("uio", "oeiu"))  # e
print(extra_letter_fast("fe", "efo"))    # o
print(extra_letter_fast("ab", "ab"))     # ""
print(extra_letter_fast("bbb", "bbbb"))  # b
```

```python
#Сумма двух элементов массива

def two_sum(nums, target):
    cache = {}
    for i, num in enumerate(nums):
        diff = target - num
        if diff in cache:
            return [cache[diff], i]
        cache[num] = i
    return []

arr = [20, 1, 2,100 ]
target = 2
print(two_sum(arr, target))
```

```python
#Сортировка Шелла

def shell_sort(arr):
    n = len(arr)
    gap = n // 2
    while gap > 0:
        for i in range(gap, n):
            temp = arr[i]
            j = i
            while j >= gap and arr[j - gap] > temp:
                arr[j] = arr[j - gap]
                j -= gap
            arr[j] = temp
        gap //= 2
    return arr

arr = [23, 12, 1, 8, 34, 54, 2, 3]
print(shell_sort(arr))
```

```python
#Массив анаграмм 

from collections import defaultdict
def group_anagrams(strs):
    groups = defaultdict(list)
    for word in strs:
        key = ''.join(sorted(word))
        groups[key].append(word)
    return list(groups.values())

print(group_anagrams(["eat", "tea", "tan", "ate", "nat", "bat"]))
print(group_anagrams(["won", "now", "aaa", "ooo", "ooo"]))
```
